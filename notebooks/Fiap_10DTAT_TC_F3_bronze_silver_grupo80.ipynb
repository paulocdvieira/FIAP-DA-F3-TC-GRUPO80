```python
lista09 =  df_092020.columns
lista10 =  df_102020.columns
lista11 = df_112020.columns
```

```python
# Converta as listas para sets para operações de conjunto rápidas
set1 = set(lista09)
set2 = set(lista10)
set3 = set(lista11)

# 1. Colunas que se repetem nas 3 listas (Interseção)
colunas_comuns = set1.intersection(set2, set3)
# Alternativa: colunas_comuns = set1 & set2 & set3

# 2. Colunas Únicas (os que aparecem em APENAS uma das listas)
# 2a. Encontre a União total de todos as colunas
uniao_total = set1 | set2 | set3

# 2b. Encontre as colunas que aparecem em EXATAMENTE duas ou três listas
comuns_em_dois_ou_mais = (set1 & set2) | (set1 & set3) | (set2 & set3)

# 2c. Subtraia as que aparecem em dois ou mais da União total
colunas_exclusivas = uniao_total.difference(comuns_em_dois_ou_mais)
# Alternativa mais simples para 'exclusivos':
# exclusivos_l1 = set1 - (set2 | set3)
# exclusivos_l2 = set2 - (set1 | set3)
# exclusivos_l3 = set3 - (set1 | set2)
# colunas_exclusivas = exclusivos_l1 | exclusivos_l2 | exclusivos_l3

print(f"Colunas Comuns nas 3 listas: {colunas_comuns}")
print(f"Colunas Exclusivos (aparecem em apenas 1 lista): {colunas_exclusivas}")
```

```python
# As bases 09 e 10 tem as mesmas colunas, a base 11 tem 3 colunas distintas A006A,A006B e A0074

exclusivos_l09 = set1 - (set2 | set3)
exclusivos_l10 = set2 - (set1 | set3)
exclusivos_l11 = set3 - (set1 | set2)


print(f"Exclusivos da Lista 1: {exclusivos_l09}")
print(f"Exclusivos da Lista 2: {exclusivos_l10}")
print(f"Exclusivos da Lista 3: {exclusivos_l11}")
```

Transformando a coluna uf de códigos para nome dos estados

```python
map_uf ={
   "11": 'Rondônia',
   "12": 'Acre',
   "13": 'Amazonas',
   "14": 'Roraima',
   "15": 'Pará',
   "16": 'Amapá',
   "17": 'Tocantins',
   "21": 'Maranhão',
   "22": 'Piauí',
   "23": 'Ceará',
   "24": 'Rio Grande do Norte',
   "25": 'Paraíba',
   "26": 'Pernambuco',
   "27": 'Alagoas',
   "28": 'Sergipe',
   "29": 'Bahia',
   "31": 'Minas Gerais',
   "32": 'Espírito Santo',
   "33": 'Rio de Janeiro',
   "35": 'São Paulo',
   "41": 'Paraná',
   "42": 'Santa Catarina',
   "43": 'Rio Grande do Sul',
   "50": 'Mato Grosso do Sul',
   "51": 'Mato Grosso',
   "52": 'Goiás',
   "53": 'Distrito Federal'
}
```

```python
base_silver_colunas3 = base_silver_colunas2.replace(map_uf, subset=["uf"])
```

```python
nome_coluna = "uf"

valores_unicos_df = base_silver_colunas3.select(nome_coluna).distinct()
valores_unicos_df.show(100)
```

Criando coluna Região

```python
from pyspark.sql import functions as f

# Cria colunas novas
base_silver_colunas3 = base_silver_colunas3.withColumn("regiao",
    f.when(f.col("uf").isin('São Paulo', 'Rio de Janeiro', 'Minas Gerais', 'Espírito Santo'), 'Sudeste')
     .when(f.col("uf").isin('Paraná', 'Santa Catarina', 'Rio Grande do Sul'), 'Sul')
     .when(f.col("uf").isin('Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Distrito Federal'), 'Centro-Oeste')
     .when(f.col("uf").isin('Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins'), 'Norte')
     .when(f.col("uf").isin('Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Paraíba', 'Pernambuco', 'Alagoas', 'Sergipe', 'Bahia'), 'Nordeste')
     .otherwise('Desconhecido') # Valor padrão caso não encontre correspondência
)
```

```python
base_silver_colunas3.show()
```

```python
nome_coluna = "regiao"

valores_unicos_df = base_silver_colunas3.select(nome_coluna).distinct()
valores_unicos_df.show(100)
```

Cria coluna com Resultado de Todos os Exames - "resultado_exames"

```python
base_silver_colunas3 = base_silver_colunas3.withColumn("resultado_exames",
    f.when(
        (f.col("resultado_SWAB") == 1) |
        (f.col("resultado_sangue_furo_dedo") == 1) |
        (f.col("resultado_sangue_veia_braco") == 1),
        1
    ).otherwise(2) # O Spark entende o None do Python como NULL
)
```
